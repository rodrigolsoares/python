```python
#Metodos responsáveis por extrair os arquivos csv que estão dentro do zip
import os
import zipfile

def searchFiles(*directories):
   listFiles = list();
   for item in directories:
        for p, _, files in os.walk(os.path.abspath(item)):
            for file in files:
                listFiles.append(os.path.join(p, file))
   return listFiles

def unzipFile(fileZip, outputFile):
    fantasy_zip = zipfile.ZipFile(fileZip)
    fantasy_zip.extractall(outputFile)
    fantasy_zip.close()
```

```python
#Rotina que extrai os arquivos csv que estão dentro dos arquivos zip
import csv

inputFile = './input/'
outputFile = './output/'

print('Start proccess')

print('Read files of type zip')
zipFiles = searchFiles(inputFile)

print('to extract file')
for zipFile in zipFiles:
    print(zipFile)
    unzipFile(zipFile, outputFile)
    
```
